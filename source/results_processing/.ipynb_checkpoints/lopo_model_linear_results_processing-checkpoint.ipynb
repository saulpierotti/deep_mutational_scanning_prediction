# LOPO model linear results processing

Here I put together the validation and testing results for the LOPO linear model.
There is no random search since the linear model has no hyperparameters.
I save the output as a more universal compressed csv file.

## Test set results

```python
import joblib
import pandas as pd

infile = "../../testing_results/lopo_models/linear_regression/validation_and_test_20210602-165230.joblib.xz"
outfile = "../../testing_results/lopo_models/linear_regression/test_results.csv.xz"

result_dump = joblib.load(infile)
result_dump["data"].to_csv(outfile, compression="xz", index=False)
pd.read_csv(outfile)
```

## Feature importance

```python
import joblib
import pandas as pd

infile = "../../testing_results/lopo_models/linear_regression/validation_and_test_20210609-154800.joblib.xz"
outfile = "../../testing_results/lopo_models/linear_regression/feature_importance.csv.xz"

result_dump = joblib.load(infile)
result_dump["feature_importance"].to_csv(outfile, compression="xz", index=False)
pd.read_csv(outfile)
```
