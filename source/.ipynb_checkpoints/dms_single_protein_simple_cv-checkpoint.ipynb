## Plotting the results with a specific set of hyperparams

```python
import itertools
import os

import joblib
import numpy as np
import pandas as pd
import xgboost as xgb
from scipy import stats
from sklearn.model_selection import KFold
from sklearn.utils import shuffle
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt


def get_model(x_train, y_train, params):
    """
    Train an instance of the XGB predictor using params, and the
    data x and y. The model is returned.
    """
    possible_xgb_params = [
        "booster",
        "verbosity",
        "validate_parameters",
        "nthread",
        "disable_default_eval_metric",
        "num_pbuffer",
        "num_feature",
        "eta",
        "gamma",
        "max_depth",
        "min_child_weight",
        "max_delta_step",
        "subsample",
        "sampling_method",
        "colsample_bytree",
        "lambda",
        "alpha",
        "tree_method",
        "sketch_eps",
        "scale_pos_weight",
        "updater",
        "refresh_leaf",
        "process_type",
        "grow_policy",
        "max_leaves",
        "max_bin",
        "predictor",
        "num_parallel_tree",
        "single_precision_histogram",
        "deterministic_histogram",
    ]
    xgb_params = {
        key: value
        for key, value in params.items()
        if key in possible_xgb_params
    }
    d_train = xgb.DMatrix(data=x_train, label=y_train)
    model = xgb.train(
        xgb_params, d_train, num_boost_round=params["num_rounds"]
    )

    return model



def get_val_pred(x_train, x_val, y_train, params):
    """
    Get the validation predictions for a given set of params
    """
    model = get_model(x_train, y_train, params)
    d_val = xgb.DMatrix(data=x_val)

    return model.predict(d_val)


def get_x_y(df_dms, features, label):
    """
    Extracts features and labels from the dataframe used for the dms study
    """

    # get dummies is an amazing function that automagically converts to 1-hot
    # the categorical columns, giving sensible names to the new columns
    x_dms = pd.get_dummies(df_dms[features]).to_numpy(dtype="float")
    y_dms = df_dms[label].to_numpy(dtype="float")

    return x_dms, y_dms


def train_test_split(x_all, y_all, is_wt_all):
    """
    Split x_all and y_all in equal train and test splits avoiding to place
    instances from the same group in both splits.
    """
    x_all_shuffled, y_all_shuffled, is_wt_all_shuffled = shuffle(  # type: ignore
        x_all, y_all, is_wt_all, random_state=1
    )
    train, test = next(
        KFold(n_splits=2).split(
            x_all_shuffled, y_all_shuffled
        )
    )

    return (
        x_all_shuffled[train],
        x_all_shuffled[test],
        y_all_shuffled[train],
        y_all_shuffled[test],
        is_wt_all_shuffled[train],
        is_wt_all_shuffled[test],
    )

def get_val_pred_val_true_cv(x_cv, y_cv, params, is_wt_cv):
    """
    Perform a 5-fold cross validation while respecting groups_cv. Returns the
    predictions on the validation sets as a concatenated array. Returns also
    the true labels in the same order.
    """
    y_val_pred = []
    y_val_true = []
    is_wt_val = []

    for train, val in KFold(n_splits=5).split(
        x_cv,
    ):
        y_val_pred.append(
            get_val_pred(
                x_cv[train],
                x_cv[val],
                y_cv[train],
                params,
            )
        )
        is_wt_val.append(is_wt_cv[val])
        y_val_true.append(y_cv[val])

    return np.concatenate(y_val_pred), np.concatenate(y_val_true), np.concatenate(is_wt_val)


def get_test_pred(x_cv, y_cv, x_test, y_test, params):
    """
    Train on x_cv and return arrays with true and predicted labels for x_test
    """
    model = get_model(x_cv, y_cv, params)
    d_test = xgb.DMatrix(data=x_test)
    
    return model.predict(d_test)

def plot_results(y_pred, y_true, is_wt, curr_dataset, kind):
    print('{} performance for dataset: {}'.format(kind, curr_dataset))
    print('test pearson:', stats.pearsonr(y_pred, y_true)[0])
    print('test spearman:', stats.spearmanr(y_pred, y_true)[0])
    print('test kendall:', stats.kendalltau(y_pred, y_true)[0])
    plt.close()
    sns.scatterplot(x=y_true[~is_wt],
                    y=y_pred[~is_wt],
                    marker='x')
    sns.scatterplot(x=y_true[is_wt],
                    y=y_pred[is_wt],
                    s=100,
                    alpha=0.5)
    plt.show()

def main(data_dict, params):
    x_all, y_all = get_x_y(
        data_dict["df"], data_dict["features"], data_dict["label"]
    )

    for curr_dataset in sorted(
        set(data_dict["df"][data_dict["dataset"]])
    ):
        x_protein = x_all[
                data_dict["df"][data_dict["dataset"]] == curr_dataset
            ]
        y_protein = y_all[
                data_dict["df"][data_dict["dataset"]] == curr_dataset
            ]
        is_wt_protein = (data_dict["df"][
                data_dict["df"][data_dict["dataset"]] == curr_dataset
            ].aa1 == data_dict["df"][
                data_dict["df"][data_dict["dataset"]] == curr_dataset
            ].aa2).to_numpy(dtype=bool)
        x_cv, x_test, y_cv, y_test, is_wt_cv, is_wt_test = train_test_split(
            x_protein,
            y_protein,
            is_wt_protein,
        )

        y_val_pred, y_val_true, is_wt_val = get_val_pred_val_true_cv(x_cv, y_cv, params, is_wt_cv)
        y_test_pred = get_test_pred(x_cv, y_cv, x_test, y_test, params)
        y_test_true = y_test
        plot_results(y_val_pred, y_val_true, is_wt_val, curr_dataset, "Validation")
        plot_results(y_test_pred, y_test_true, is_wt_test, curr_dataset, "Test")

DATA_DICT = {
    # This is the main dataframe with all the features and response variables,
    # and also additional information
    "df": pd.read_csv(
        "/home/saul/master_thesis_work/dataset/dms/dms_training.csv",
    ),
    # which column of df to use as label for the regression
    "label": "reported_fitness",
    # columns of df identifying datasets that should be processed
    # separately
    "dataset": "dms_id",
    # the features that I want my predictor to use. Comment out the ones that
    # need to be excluded
    "features": [
        "aa1",
        "aa2",
        "ev_frequency",
        "ev_conservation",
        "ev_independent",
        "ev_epistatic",
        "netsurf_rsa",
        "netsurf_asa",
        "netsurf_p_q3_H",
        "netsurf_p_q3_E",
        "netsurf_p_q3_C",
        "netsurf_p_q8_G",
        "netsurf_p_q8_H",
        "netsurf_p_q8_I",
        "netsurf_p_q8_B",
        "netsurf_p_q8_E",
        "netsurf_p_q8_S",
        "netsurf_p_q8_T",
        "netsurf_p_q8_C",
        "netsurf_phi",
        "netsurf_psi",
        "netsurf_disorder",
        "hmm_pssm_A",
        "hmm_pssm_C",
        "hmm_pssm_D",
        "hmm_pssm_E",
        "hmm_pssm_F",
        "hmm_pssm_G",
        "hmm_pssm_H",
        "hmm_pssm_I",
        "hmm_pssm_K",
        "hmm_pssm_L",
        "hmm_pssm_M",
        "hmm_pssm_N",
        "hmm_pssm_P",
        "hmm_pssm_Q",
        "hmm_pssm_R",
        "hmm_pssm_S",
        "hmm_pssm_T",
        "hmm_pssm_V",
        "hmm_pssm_W",
        "hmm_pssm_Y",
        "hmm_pssm_aa1_likelyhood",
        "hmm_pssm_aa2_likelyhood",
        "hmm_pssm_delta_likelyhood",
        "tr_rosetta_graph_closeness_centrality",
        "tr_rosetta_graph_betweenness_centrality",
        "tr_rosetta_graph_degree_centrality",
        "tr_rosetta_graph_load_centrality",
        "tr_rosetta_graph_harmonic_centrality",
        "tr_rosetta_graph_clustering",
    ],
}

PARAMS = {'max_depth': 2,
 'min_child_weight': 1,
 'subsample': 0.4,
 'colsample_bytree': 0.2,
 'eta': 0.03,
 'tree_method': 'hist',
 'objective': 'reg:squarederror',
 'nthread': -1,
 'gamma': 0.005,
 'lambda': 0.01,
 'alpha': 0.01,
 'num_rounds': 1000}

main(DATA_DICT, PARAMS)
```
