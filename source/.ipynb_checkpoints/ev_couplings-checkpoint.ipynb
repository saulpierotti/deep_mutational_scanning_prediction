# EV couplings

Here I detailed the pipeline for installing and running the EV coupling package. The package can produce alignments, predict folding and predict the effect of mutations using an independent or epistatic unsupervised model.

## Installation

EV coupling is a pip package for python 3.5+

```
python3 -m virtualenv -p3 ev_couplings_env
source ev_couplings_env/bin/activate
pip install evcouplings
```

EV Couplings requires plmc (same lab) compiled with `make all-openmp32`. Plmc is a tool for inferring undirected statistical models from sequence variation. Plmc has no dependecies.

```
git clone git@github.com:debbiemarkslab/plmc.git
cd plmc
make all-openmp32
```

EV Couplings requires HMMER. It is available in the AUR but I need to run on the server since the databases are heavy. On the server it is already installed.

```
yay -S hmmer
```

EV couplings requires optionally hhblits (already installed on the server and also locally), CNS (for modelling, not needed in this case), psipred (for folding, not needed), maxcluster (for folding, not needed).

## Running

EV coupling is configured with a text file and optional parameters can be passed from the command line. I use the following configuration script

````python
!cat ../processing/gray2018/ev_couplings/config_aeserv19a.txt
````

I run the actual computation with the followig command.

```
source ev_couplings_env/bin/activate
evcouplings -P <uniprot_id> -p <uniprot_id> -s ../uniprot_sequences/<uniprot_id>.fasta config_aeserv19a.txt 
```

I am actually interested only in retaining the MSAs in `<uniprot_id>/align/<uniprot_id>.a2m` and the mutation scores in `<uniprot_id>/mutate/<uniprot_id>_single_mutant_matrix.csv`. I can trow away the rest of the output.

## Correlation with the DMS scores

````python
import pandas as pd
import numpy as np
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
from scipy import stats

df = pd.read_csv('/home/saul/master_thesis_work/dataset/dms/dms_training.csv')

for study in set(df.dms_id):
    basenames = set(df[df.dms_id == study].feature_basename)
    assert len(basenames) == 1
    basename = basenames.pop()
    x = np.array(df[df.dms_id == study].reported_fitness)
    is_wt = (df[df.dms_id == study].aa1 == df[df.dms_id == study].aa2)
    y_independent = np.array(df[df.dms_id == study].ev_independent)
    y_epistatic = np.array(df[df.dms_id == study].ev_epistatic)
    # there are no WT in the ev couplings output so I cannot plot them!
    print(basename, study, 'Independent model')
    print("Spearman:", stats.spearmanr(x[~np.isnan(y_independent)], y_independent[~np.isnan(y_independent)]))
    plt.close()
    sns.scatterplot(x=x[~np.isnan(y_independent)], y=y_independent[~np.isnan(y_independent)], marker='x')
    plt.show()
    print(basename, study, 'Epistatic model')
    print("Spearman:", stats.spearmanr(x[~np.isnan(y_epistatic)], y_epistatic[~np.isnan(y_epistatic)]))
    plt.close()
    sns.scatterplot(x=x[~np.isnan(y_epistatic)], y=y_epistatic[~np.isnan(y_epistatic)], marker='x')
    plt.show()
````
