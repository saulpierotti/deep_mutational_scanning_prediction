# Hhblits pssm

Here I document how I obtained the MSA with hhblits and how from those I got a PSSM

## Running hhblits

Hhblits must be run on the server since the database is very large and does not fit locally. I use default parameters and a minimum coverage of 70%.

```
hhblits -i <input_seq> -o <result_file> -d <database> -oa3m <out_msa_file> -cov 70

reformat.pl a3m fas '*.a3m' '.aln' -r
```

## Getting the PSSM

In order to get a PSSM I first create an HMM with hmmbuild of the hmmer suite. I use hmmbuild and not hhmake (from hh suite) since the latter does not include pseudocounts in the emission probabilities.

```
hmmbuild <uniprot_id>.hmm ../hhblits/<uniprot_id>.aln
```

I obtain then the PSSM from the HMM by mapping the emission probabilities to the correct MSA positions. I use the match state and insert state emissions as appropriate for the position.

```
hmm_to_pssm.py -i <uniprot_id>.hmm -o <uniprot_id>.hmm_pssm.joblib.xz -m ../uniprot_sequences/<uniprot_id>.fasta 
```
