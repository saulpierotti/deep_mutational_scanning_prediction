# Feature distribution Rebb2020 detrimental

This notebook is to get a first glance at the distribution of the features and of the scores.

```python
import numpy as np
import pandas as pd; pd.set_option('display.max_rows', None)
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()

norm_score = np.load('../processing/training_set/norm_score.npy')
dataset_vec = np.load('../processing/training_set/dataset_vec.npy').flatten()
shannon_sliding_windows = np.load('../processing/training_set/nr_db/shannon_sliding_windows.npz')['arr_0']
profile_sliding_windows = np.load('../processing/training_set/nr_db/profiles_sliding_windows.npz')['arr_0']
```

The Shannon's entropy ranges from 0 to around 3, with most scores near 0 and then a pretty constant distribution.
The 0 values are due to the padding on the sliding window.

```python
df = pd.DataFrame(shannon_sliding_windows.flatten()).melt()
sns.displot(df, x='value', stat='probability')
```

Most profile positions are at 0 almost exponentially decreasing towards 0.8. Then the frequency increases from 0.8 to 1

```python
df = pd.DataFrame(profile_sliding_windows.flatten()).melt()
df = df[df['value'] > 0.1]
sns.histplot(df, x='value', stat='probability')
```

The target values, the norm score, is distributed between 0 and 1

```python
df = pd.DataFrame(norm_score).melt()
sns.histplot(df, x='value', stat='probability')
```
