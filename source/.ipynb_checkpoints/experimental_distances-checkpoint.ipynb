# Experimental distances
A notebook to create an experimental distance matrix for the training DMS studies.

PDB IDs and chains are inserted manually since I need to check which structure is best.

```python
import pandas as pd

dms_datasets_df = pd.read_csv('/home/saul/master_thesis_work/dataset/dms/dms_datasets.csv')
dms_datasets_df
```

Now I download the relevant structures and create a simple csv file ot use for extracting the contact maps (I need it as an input of one of my scripts)

```python
pdb_chains_df = dms_datasets_df[['pdb_id', 'pdb_chain']]
pdb_chains_df.to_csv('/home/saul/master_thesis_work/processing/dms/structures/pdb_chains.csv', index=False, header=False)
```

I download the structures

```python
!cd /home/saul/master_thesis_work/processing/dms/structures &&\
cat /home/saul/master_thesis_work/processing/dms/structures/pdb_chains.csv|\
cut -d, -f1|\
sort|uniq|\
xargs -I{}\
sh -c "pdb_get.sh {}"
```

## Creation of the distance matrix
The initial distance matrix extraction is obtained with a custom python script.

```python
!cd /home/saul/master_thesis_work/processing/dms/structures &&\
cat /home/saul/master_thesis_work/processing/dms/structures/pdb_chains.csv|\
sort|uniq|\
tr ',' '\n'|\
xargs -n2\
sh -c 'pdb_to_distance_matrix.py -i $0.cif -o $0_$1.pdb_distance_matrix.joblib.xz -c $1'
```

The output for each PDB file is a joblib dump of a dictionary containing the distance matrix itself as a numpy array, the primary sequence of the PDB structure, the seqres identifiers, and the 3-letter codes for each residue of the sequence.

## Mapping of the distance matrix to the Uniprot sequence

Now I map the distance matrix to the Uniprot sequence so that the numpy array has dimension $L^2$ where L is the lenght of the uniprot sequence. Missing positions in the PDB correspond to NaN entries. I use SIFTS mappings to relate PDB positions to uniprot positions. I also do this with a custom script. Some structures can be mapped using the RES field, some structures using the PDB field of the sifts mappings.

```python
!cd /home/saul/master_thesis_work/processing/dms/structures &&\
cat /home/saul/master_thesis_work/processing/dms/structures/pdb_chains.csv|\
sort|uniq|\
tr ',' '\n'|\
xargs -n2\
sh -c 'echo ;map_distance_matrix_to_uniprot.py\
-i $0_$1.pdb_distance_matrix.joblib.xz\
-db /home/saul/master_thesis_work/dataset/sifts_mappings/uniprot_segments_observed.csv.gz'
```

## Visualization of the distance matrices before and after mapping

The mapping that I did seems correct. For some protein the trimmed contact map is empty since the structures available refer to another identical domain in the same sequence. When comparing with the trRosetta output, it is better to use just the PDB unmapped contact map as a visual reference.

```python
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
import joblib

for study in set(dms_datasets_df.dms_id):
    assert len(dms_datasets_df[dms_datasets_df.dms_id == study]) == 1
    pdb_id = set(dms_datasets_df[dms_datasets_df.dms_id == study].pdb_id).pop()
    pdb_chain = set(dms_datasets_df[dms_datasets_df.dms_id == study].pdb_chain).pop()
    uniprot_id = set(dms_datasets_df[dms_datasets_df.dms_id == study].uniprot_id).pop()
    uniprot_first_idx = set(dms_datasets_df[dms_datasets_df.dms_id == study].uniprot_first).pop() - 1
    uniprot_last_idx = set(dms_datasets_df[dms_datasets_df.dms_id == study].uniprot_last).pop()
    print('PDB distance matrix', study, uniprot_id, pdb_id, pdb_chain)
    mat = joblib.load('/home/saul/master_thesis_work/processing/dms/structures/' +
                      pdb_id +'_'+ pdb_chain +
                      '.pdb_distance_matrix.joblib.xz')['distance_matrix']
    plt.close()
    sns.heatmap(mat)
    plt.show()
    print('Uniprot distance matrix', study, uniprot_id, pdb_id, pdb_chain)
    mat = joblib.load('/home/saul/master_thesis_work/processing/dms/structures/' +
                      uniprot_id + '_mapped_' + pdb_id + '_' + pdb_chain +
                      '.uniprot_distance_matrix.joblib.xz')['distance_matrix']
    plt.close()
    sns.heatmap(mat)
    plt.show()
    print('Uniprot trimmed distance matrix', study, uniprot_id, pdb_id, pdb_chain)
    print("Trimmed to", uniprot_first_idx, uniprot_last_idx)
    plt.close()
    sns.heatmap(mat[uniprot_first_idx : uniprot_last_idx,
                    uniprot_first_idx : uniprot_last_idx])
    plt.show()
```
