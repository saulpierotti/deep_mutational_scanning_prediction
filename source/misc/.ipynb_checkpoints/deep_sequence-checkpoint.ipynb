````python
from theano import function, config, shared, tensor
import numpy
import time

vlen = 10 * 30 * 768  # 10 x #cores x # threads per core
iters = 1000

rng = numpy.random.RandomState(22)
x = shared(numpy.asarray(rng.rand(vlen), config.floatX))
f = function([], tensor.exp(x))
print(f.maker.fgraph.toposort())
t0 = time.time()
for i in range(iters):
    r = f()
t1 = time.time()
print("Looping %d times took %f seconds" % (iters, t1 - t0))
print("Result is %s" % (r,))
if numpy.any([isinstance(x.op, tensor.Elemwise) and
              ('Gpu' not in type(x.op).__name__)
              for x in f.maker.fgraph.toposort()]):
    print('Used the cpu')
else:
    print('Used the gpu')
````

## Training a model

Two different model classes are offered by DeepSequence, VariationalAuroencoder and VariationalAutoencoderMLE. The first uses SVI (stochastic variational inference) and the second MLE (maximum likelihood estimation). The models are defined in `model.py`, while the training loop is defined in `train.py` and `helper.py` contains functions for data pre-processing. An example of the training pipeline is available in `examples/run_svi.py` and `examples/run_mle.py`. I base my training on the svi version. There is not enough vRAM for running this locally. I am running it on the server. I created a script `server_jobs/train_deepsequence.py` to train the model on the server. I just give the ev coupling MSA as argument:

```
train_deepsequence.py <input_msa_a2m>
```

It creates a folder `logs` and a folder `params` in the current directory with the outputs. Multiple models can be written to the same folder.

### How I do training

In the server there are 2 GPUs with 8 Gb of vRAM each. I have 9 inputs and sending more than 1 input at a time on the same GPU crashes it. Therefore I send inputs 1 at a time for each GPU. I can decide which GPU to use by exporting (not just setting, EXPORTING!) the variable `$THEANO_FLAGS=device=cuda0` before running the training (`cuda1` for the other GPU). 

### Problems when training

Everything seems ok for the rest of ev_couplings the inputs in gray2018, but P31016.a2m goes out of memory (8 Gb) on the server GPU. I tried to resize it in different ways (`hhfilter -diff 1000`, `hhfilter -cov50`, `reformat.pl -r`) but nothing worked. I think it may be due to the size of the protein (~750 aa). I am therefore running it on cpu (`$THEANO_FLAGS=device=cpu`).

## Correlation with the DMS scores

````python
import pandas as pd
import numpy as np
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
from scipy import stats

df = pd.read_csv('/home/saul/master_thesis_work/dataset/dms/dms_training.csv')

for study in set(df.dms_id):
    basenames = set(df[df.dms_id == study].feature_basename)
    assert len(basenames) == 1
    basename = basenames.pop()
    x = np.array(df[df.dms_id == study].reported_fitness)
    y = np.array(df[df.dms_id == study].deepseq_score_minmax)
    print(basename, study, 'Deepsequence delta ELBO')
    print("Spearman:", stats.spearmanr(x[~np.isnan(y)], y[~np.isnan(y)]))
    plt.close()
    sns.scatterplot(x=x[~np.isnan(y)], y=y[~np.isnan(y)])
    plt.show()
````

## Example run

````python
import theano
import numpy as np
import sys
import pandas as pd
import scipy
from scipy.stats import spearmanr

%matplotlib inline
import matplotlib.pyplot as plt


deepsequence_path = '/home/saul/packages/deepsequence/DeepSequence/'

if deepsequence_path not in sys.path:
    sys.path.insert(0, deepsequence_path)

import model
import helper
import train

# now I need to change the path hardcoded in helper.py since it is set wrong
# I just remove the non-existent folder dataset from i
````

````python
working_dir = '/home/saul/packages/deepsequence/examples'
alignment_file = '/home/saul/packages/deepsequence/examples/alignments/BLAT_ECOLX_hmmerbit_plmc_n5_m30_f50_t0.2_r24-286_id100_b105.a2m'

data_params = {
    "working_dir": working_dir,
    "alignment_file": alignment_file,
}

model_params = {
    "bs"                    : 100,
    "encode_dim_zero"       : 1500,
    "encode_dim_one"        : 1500,
    "decode_dim_zero"       : 100,
    "decode_dim_one"        : 500,
    "n_latent"              : 30,
    "logit_p"               : 0.001,
    "sparsity"              : "logit",
    "final_decode_nonlin"   : "sigmoid",
    "final_pwm_scale"       : True,
    "n_pat"                 : 4,
    "r_seed"                : 12345,
    "conv_pat"              : True,
    "d_c_size"              : 40,
    "working_dir"           : working_dir,
}

data_helper = helper.DataHelper(
    working_dir=data_params["working_dir"],
    alignment_file=data_params["alignment_file"],
    calc_weights=True,
)

vae_model = model.VariationalAutoencoder(
    data_helper,
    batch_size=model_params["bs"],
    encoder_architecture=[
        model_params["encode_dim_zero"],
        model_params["encode_dim_one"],
    ],
    decoder_architecture=[
        model_params["decode_dim_zero"],
        model_params["decode_dim_one"],
    ],
    n_latent=model_params["n_latent"],
    logit_p=model_params["logit_p"],
    sparsity=model_params["sparsity"],
    encode_nonlinearity_type="relu",
    decode_nonlinearity_type="relu",
    final_decode_nonlinearity=model_params["final_decode_nonlin"],
    final_pwm_scale=model_params["final_pwm_scale"],
    conv_decoder_size=model_params["d_c_size"],
    convolve_patterns=model_params["conv_pat"],
    n_patterns=model_params["n_pat"],
    random_seed=model_params["r_seed"],
    working_dir=model_params["working_dir"],
)

print ("Model built")
````

````python
file_prefix = "BLAT_ECOLX"
vae_model.load_parameters(file_prefix=file_prefix)
print ("Parameters loaded")
````

````python
matr_mutant_name_list, matr_delta_elbos = data_helper.single_mutant_matrix(vae_model, N_pred_iterations=500)
    
print("Mutant matrix ready")
````
