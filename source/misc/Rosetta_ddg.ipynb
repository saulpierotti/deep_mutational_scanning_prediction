# Rosetta DDG
The aim of this notebook is to track the development of a pipeline for the creation of a synthetic DMS/stability dataset. The idea is to use Rosetta to estimate the DDG of a set of point mutations (very slow) and use that as training data for my implementation in order to expand from the limited DMS data. The stated correlation of the Rosetta protocol with experiments is around 0.69.

## Getting the DDG scores

### Input structure preparation

I first clean the input PDB file and select the relevant chain with the clean_pdb.py script. This separates single chains from the pdb, renumbers the residues from 1 to L in the way that Rosetta expects, and strips away unrecognised residues and all non-ATOM rows. The input is a PDB file and the output is a PDB file and a fasta file with the corresponding sequence.

```
python2 rosetta_src_2021.07.61567_bundle/main/tools/protein_tools/scripts/clean_pdb.py <pdb_file> <chain_id(s)>
```

I then relax the structure to minimize the energy function without altering too much the structure. This is so that I can do a fair comparison between the relaxed WT and the relaxed mutant. I use for now the relax:fast protocol since it is much faster and should have little impact on accuracy.

```
rosetta_src_2021.07.61567_bundle/main/source/bin/relax.default.linuxgccrelease -relax:fast -in:file:s <pdb_file>
```

### Generation of the mutation list

I now define which mutations should be applied. To do this, I use the fasta sequence from clean_pdb to determine the WT residues. This information is to be given to Rosetta as a file with the following format:

```
total 2
1
D 1 A
```

`total n` must represent the total number of mutations to make. Each mutation should be specified separately otherwise they are made sequentially on the same model. Each mutation group should be preceded by a number indicating how many mutations are in the group (1 for all groups). The actual mutation line specifies `wt_res seqres mut_res`

````python
from Bio import SeqIO, Data

#alphabet = set(Data.IUPACData.protein_letters)
# for now I try an alanine scanning
alphabet = set('A')
fasta_in = '../processing/gray2018/rosetta_ddg/cleaned/1nd4_A.fasta'
mut_file_out = '../processing/gray2018/rosetta_ddg/ddg_monomer/1nd4_A.mutlist.txt'
seq = SeqIO.read(fasta_in, 'fasta').seq
n_muts_tot = len(seq)*(len(alphabet))
out_mut_str = 'total ' + str(n_muts_tot) + '\n'
mut_count_check = 0
for i, wt_res in enumerate(seq):
    # I do also same mutations so to have a score also for them
    for mut_res in alphabet:
        out_mut_str += ('1\n' + wt_res + ' ' + str(i+1) + ' ' + mut_res + '\n')
        mut_count_check += 1
assert mut_count_check == n_muts_tot
with open(mut_file_out, 'w') as handle:
    handle.write(out_mut_str)
````

### Rosetta modelling

It seems to do very little repacking outside of the mutated position, my guess is that it is protocol 1.

## Comparing the DDG scores to the DMS scores

````python
import pandas as pd
from scipy import stats
import numpy as np

ddg_output_file = '../processing/gray2018/rosetta_ddg/ddg_monomer/ddg_predictions.out'
ddg_df = pd.read_csv(ddg_output_file, header=0, sep='\s+', usecols=range(1,3), names=['mut_rosetta', 'ddg_rosetta'])
ddg_df['score_rosetta'] = - ddg_df.ddg_rosetta
ddg_df['aa1'] = ddg_df.mut_rosetta.apply(lambda x : x[0])
ddg_df['aa2'] = ddg_df.mut_rosetta.apply(lambda x : x[-1])
ddg_df['position_rosetta'] = ddg_df.mut_rosetta.apply(lambda x : int(x[1:-1]))
ddg_df['position'] = ddg_df.position_rosetta + 9 #only for this, find a better way!
ddg_df['uniprot_id'] = 'P00552'

dms_df = pd.read_csv('../dataset/gray2018/dmsTraining_2017-02-20.csv')

df = pd.merge(ddg_df, dms_df)[['dms_id', 'protein', 'pdb_id', 'uniprot_id', 'position', 'position_rosetta', 'aa1', 'aa2', 'mut_rosetta', 'reported_fitness', 'score_rosetta']]


print('Pearson:', stats.pearsonr(df.reported_fitness, df.score_rosetta))
print('Spearman:', stats.spearmanr(df.reported_fitness, df.score_rosetta))
````
