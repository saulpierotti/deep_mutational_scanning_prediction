# Feature correlation Rebb2020 detrimental

This notebook is to get a first glance of the correlation of various features with the measured mutation effect.

```python
import numpy as np
import pandas as pd; pd.set_option('display.max_rows', None)
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
from scipy import stats

norm_score = np.load('../processing/training_set/norm_score.npy')
dataset_vec = np.load('../processing/training_set/dataset_vec.npy').flatten()
shannon_sliding_windows = np.load('../processing/training_set/nr_db/shannon_sliding_windows.npz')['arr_0']
profile_sliding_windows = np.load('../processing/training_set/nr_db/profiles_sliding_windows.npz')['arr_0']
mutation_vec = np.load('../processing/training_set/mutations_two_hot.npy')
```

I create a single feature array and format correctly the target array. I also create a randomized output vector to compare the correlations. I also produce the cv LOPO splits.

```python
def get_window_range(sliding_window_array):
    window_size = sliding_window_array.shape[1]
    upper = ((window_size - 1) // 2) + 1
    lower = - ((window_size - 1) // 2)
    return range(lower, upper)

# get the indeces for the sliding windows positions
shannon_window_range = get_window_range(shannon_sliding_windows)
profile_window_range = get_window_range(profile_sliding_windows)

# reshape the vector so that it is 1d
y = norm_score.flatten()

# create the input vector by concatenating the features and flattening them
x = np.concatenate([shannon_sliding_windows.reshape(len(shannon_sliding_windows),-1),
                    profile_sliding_windows.reshape(len(profile_sliding_windows),-1),
                    mutation_vec,
                   ],
                   axis=1)

# obtain a vector with the names of the features for easier recall
x_features = np.array(['shannon_' + str(pos) for pos in shannon_window_range] +
              ['profile_' + str(aa) + '_' + str(pos) for aa in range(20) for pos in profile_window_range] +
              ['mutation_' + kind + '_' + str(aa) for kind in ('from', 'to') for aa in range(20)])

# a boolean vector that differentiates numerical and categorical features
x_is_numerical = np.array([True] * shannon_sliding_windows.shape[1] +
                  [False] * profile_sliding_windows.shape[1] * 20 +
                  [False] * mutation_vec.shape[1]
                         )

# slicing the input vectors so to separate numerical and categorical features
x_num = x.T[x_is_numerical].T
x_cat = x.T[np.logical_not(x_is_numerical)].T

# obtaining the feature name vector for the numerical and categorical sub-inputs
x_num_features = x_features.T[x_is_numerical].T
x_cat_features = x_features.T[np.logical_not(x_is_numerical)].T
```

## Numerical featurs

I look at the correalation of the numerical features with the numerical output. These are expressed in terms of Pearson's and Spearman's correlation.

```python
def plot_numerical_feature_correlation(x, y):
    
    # get a shuffled copy of y to compare correlations against
    ran_y = y.copy()
    np.random.shuffle(ran_y)
    
    # calculate the correlations and put them in a dataframe for plotting
    df = []
    for column in x.T: # for each feature type
        spearman = stats.spearmanr(column, y)[0]
        pearson = stats.pearsonr(column, y)[0]
        random_spearman = stats.spearmanr(column, ran_y)[0]
        random_pearson = stats.pearsonr(column, ran_y)[0]
        df.append([spearman, pearson, random_spearman, random_pearson])
    df = pd.DataFrame(df, columns=['spearman', 'pearson', 'random_spearman', 'random_pearson'], index=x_num_features)
    
    # draw the plot
    plt.close()
    plt.plot(list(shannon_window_range), df['spearman'], label = 'spearman')
    plt.plot(list(shannon_window_range), df['pearson'], label = 'pearson')
    plt.plot(list(shannon_window_range), df['random_spearman'], label = 'random_spearman')
    plt.plot(list(shannon_window_range), df['random_pearson'], label = 'random_pearson')
    plt.legend()
    
    # I I want to zoom in a specific region
    # plt.xlim([-500,500])
    
    plt.show()
```

### Shannon's entropy

Shannon is reported on a sliding window. Here I plot the correlation of each position along the window with the score of the mutation. There is some weak correlation. The central position of the window is the one with the strongest correlation, as expected. The correlation is negative as expected for the central position.

```python
plot_numerical_feature_correlation(x_num, y)
```

I do the same now but separating by dataset, to see if there are particularly bad behaved datasets in the ensemble.

```python
datasets = np.unique(dataset_vec)

for dataset in datasets:
    print(dataset)
    bool_vec = (dataset_vec == dataset).flatten()
    curr_x, curr_y = x_num[bool_vec], y[bool_vec]
    plot_numerical_feature_correlation(curr_x, curr_y)
```
